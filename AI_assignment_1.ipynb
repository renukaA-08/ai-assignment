```python
class State:
    def __init__(self, config):
        self.config = config 

    def goalTest(self):
        return self.config == ['W', 'W', 'W', '_', 'E', 'E', 'E']

    def moveGen(self):
        children = []
        for i in range(len(self.config)):
            if self.config[i] == 'E':
                if i+1 < 7 and self.config[i+1] == '_':
                    new_config = self.config[:]
                    new_config[i], new_config[i+1] = new_config[i+1], new_config[i]
                    children.append(State(new_config))
                if i+2 < 7 and self.config[i+2] == '_' and self.config[i+1] in ['E', 'W']:
                    new_config = self.config[:]
                    new_config[i], new_config[i+2] = new_config[i+2], new_config[i]
                    children.append(State(new_config))
            elif self.config[i] == 'W':
                if i-1 >= 0 and self.config[i-1] == '_':
                    new_config = self.config[:]
                    new_config[i], new_config[i-1] = new_config[i-1], new_config[i]
                    children.append(State(new_config))
                if i-2 >= 0 and self.config[i-2] == '_' and self.config[i-1] in ['E', 'W']:
                    new_config = self.config[:]
                    new_config[i], new_config[i-2] = new_config[i-2], new_config[i]
                    children.append(State(new_config))
        return children

    def __str__(self):
        return ''.join(self.config)

    def __eq__(self, other):
        return self.config == other.config

    def __hash__(self):
        return hash(tuple(self.config))

def bfs(start_state):
    OPEN = []
    CLOSED = set()
    OPEN.append((start_state, [start_state]))

    while OPEN:
        current, parent = OPEN.pop(0)
        if current in CLOSED:
            continue
        CLOSED.add(current)

        if current.goalTest():
            return parent
        else:
            #children=current.moveGen()
            for child in current.moveGen():
                if child not in CLOSED:
                    OPEN.append((child, parent + [child]))
    return None

def dfs(state, CLOSED, parent):
    if state in CLOSED:
        return None
    CLOSED.add(state)

    if state.goalTest():
        return parent + [state]

    for child in state.moveGen():
        result = dfs(child, CLOSED, parent + [state])
        if result:
            return result
    return None

start_state = State(['E', 'E', 'E', '_', 'W', 'W', 'W'])

print(" BFS Solution Path:")
bfs_solution = bfs(start_state)
for i, step in enumerate(bfs_solution):
    print(f"Step {i}: {step}")

print("\n DFS Solution Path:")
dfs_solution = dfs(start_state, CLOSED=set(), parent=[])
if dfs_solution:
    for i, step in enumerate(dfs_solution):
        print(f"Step {i}: {step}")
else:
    print("No solution found using DFS.")
```

```python
class State:
    def __init__(self, amogh, ameya, grandma, grandpa, umbrella, time):
        self.amogh = amogh
        self.ameya = ameya
        self.grandma = grandma
        self.grandpa = grandpa
        self.umbrella = umbrella
        self.time = time

    def isGoal(self):
        return (self.amogh == 'R' and self.ameya == 'R' and
                self.grandma == 'R' and self.grandpa == 'R' and
                self.time <= 60)

    def __eq__(self, other):
        return (self.amogh == other.amogh and self.ameya == other.ameya and
                self.grandma == other.grandma and self.grandpa == other.grandpa and
                self.umbrella == other.umbrella)

    def __hash__(self):
        return hash((self.amogh, self.ameya, self.grandma, self.grandpa, self.umbrella))

    def __str__(self):
        return f"Amo:{self.amogh} Ame:{self.ameya} Gma:{self.grandma} Gpa:{self.grandpa} Umb:{self.umbrella} Time:{self.time}"

    def moveGen(self):
        times = {'Amogh': 5, 'Ameya': 10, 'Grandma': 20, 'Grandpa': 25}
        sides = {
            'Amogh': self.amogh,
            'Ameya': self.ameya,
            'Grandma': self.grandma,
            'Grandpa': self.grandpa
        }

        curr_side = self.umbrella
        otherside = 'R' if curr_side == 'L' else 'L'
        people_here = [p for p in sides if sides[p] == curr_side]

        children = []

        for i in range(len(people_here)):
            for j in range(i, len(people_here)):
                p1, p2 = people_here[i], people_here[j]

                new_sides = sides.copy()
                new_sides[p1] = otherside
                new_sides[p2] = otherside

                time_needed = max(times[p1], times[p2])
                new_time = self.time + time_needed

                child = State(
                    amogh=new_sides['Amogh'],
                    ameya=new_sides['Ameya'],
                    grandma=new_sides['Grandma'],
                    grandpa=new_sides['Grandpa'],
                    umbrella=otherside,
                    time=new_time
                )
                children.append((child, [p1, p2], time_needed))
        return children

def bfs(start):
    OPEN = [(start, [(start, [], 0)])]
    CLOSED = []

    while OPEN :
        state, path = OPEN .pop(0)
        if state in CLOSED:
            continue
        CLOSED.append(state)

        if state.isGoal():
            return path

        for child, movers, cost in state.moveGen():
            if child.time <= 60 and child not in CLOSED:
                OPEN .append((child, path + [(child, movers, cost)]))
    return None
    
def dfs(state, path,CLOSED):
    if state.isGoal():
        return path
    CLOSED.append(state)

    for child, movers, cost in state.moveGen():
        if child.time <= 60 and child not in CLOSED:
            res = dfs(child, path + [(child, movers, cost)], CLOSED)
            if res:
                return res
    return None
    
start = State('L', 'L', 'L', 'L', 'L', 0)

print("=== BFS Solution ===")
bfs_solution = bfs(start)
if bfs_solution:
    for i, (s, movers, cost) in enumerate(bfs_solution):
        if i == 0:
            print(f"Start: {s}")
        else:
            print(f"Step {i}: {movers} cross --> +{cost} mins --> {s}")
else:
    print("No BFS solution found.")

print("\n=== DFS Solution ===")
dfs_solution = dfs(start, [(start, [], 0)], [])
if dfs_solution:
    for i, (s, movers, cost) in enumerate(dfs_solution):
        if i == 0:
            print(f"Start: {s}")
        else:
            print(f"Step {i}: {movers} cross --> +{cost} mins --> {s}")
else:
    print("No DFS solution found.")
```
